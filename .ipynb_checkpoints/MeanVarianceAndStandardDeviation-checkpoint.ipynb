```python
# manual scaling notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

# lib used to standrdize data, aldoh in notebook we will try doing this manually
from sklearn.preprocessing import StandardScaler, scale
```

```python
df = pd.read_csv("./Datasets/Absenteeism_at_work.csv", delimiter=";")

df.head()
```

```python
# gets col names of features in df
df.columns
```

```python
# select usefui cols for us to work with
df = df[['Transportation expense',
        'Distance from Residence to Work',
        'Service time',
        'Age',
        'Work load Average/day ',
        'Weight',
        'Height',
        'Body mass index',
        'Absenteeism time in hours']]
```

```python
# rename cols
df.rename(columns = {
    'Transportation expense': 'transportation',
    'Distance from Residence to Work': 'Distance',
    "Service time": 'service',
    'Work load Average/day ': 'workload',
    'Body mass index': 'BMI',
    'Absenteeism time in hours': 'AbsentHours'
}, inplace = True)
```

```python
df.head()
```

```python
df = df.astype(np.float64)
```

```python
print("transporatiion mean: ", np.mean(df.transportation.values))

print("transporatiion median: ", np.median(df.transportation.values))

print("transporatiion variance: ", np.var(df.transportation.values))

print("transporatiion standard deviation: ", np.std(df.transportation.values))


# alt way of getting these stuff if using a df
print("\ntransporatiion mean df: ", df.transportation.mean())

print("transporatiion medain df: ", df.transportation.median())
```

```python
# gets stat info about num features in dataFrame
df.describe()
```

```python
# data viz
df.boxplot(column=["transportation"], figsize = (8, 8))

plt.show();
```

```python
# data viz
df.boxplot(column=["transportation", "Distance"], figsize = (8, 8))


# can't compare since, visually clear that both features are on diff scales
plt.show();
```

```python
# standarzing manually(standardizing each feature singlularly)
scaled_df = pd.DataFrame()

scaled_df["Transportation"] = (df.transportation - np.mean(df.transportation.values)) / \
    np.std(df.transportation.values)

scaled_df["Distance"] = (df.Distance - np.mean(df.Distance.values)) /\
    np.std(df.Distance.values)

scaled_df["AbsentHours"] = (df.AbsentHours - np.mean(df.AbsentHours.values))  /\
    np.std(df.AbsentHours.values)
```
